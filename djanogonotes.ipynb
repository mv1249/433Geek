## Template Inheritance


**→We use the concept of template inheritance,inorder to maintain some "same components" in all pages,i.e suppose our website has 10 pages and i want all the pages to have the same header and footer throughout,so i can make a template let's say "base" and in this i'll only mention the header,navbar and footer,and let all the other 10 templates inherit the properties of the base template,so in all those 10 templates which has inherited from the base template will have the same header,navbar and fotter in them,now in those 10 templates you write your code whatever you want to write,and if suppose you want to add other 2 templates,then again no need to write the header,navbar and fotter again,just inherit the "base" template,so template inheritance basically reduces the length of code and makes teh code reusable,syntax for writing the inheritance is :**

<font color = blue >{%block body%}{% endblock body%}</font>

**→This you must write at the end of your template file,i.e after < /body >,note that the name of the block here is 'body'!**


**→for inheriting the template use this : <font color = blue >{%extends < template_name >%}</font>,here the template_name is the name of the template from which you're inherting which in this case is base.html***

**→This is the `base.html` file where in i've written the code for nav_bar and base.html act's as a parent class for all the templates!!,as you can observe that after the end of the < /nav >block we've written the {%block body}{%endblock body} which is an indication that the next template gets inherited from this parent template!!**

**This is the index.html template which is extending the base.html file,as we can see that first is the `extend block`,then is the `block` whose name is body and then there's the place where we can place the content related to the `index` block!,and then end the block! as we're extending the base.html,we have navbars and footers in all the templates!!,as shown below ↓**

<font color = red size = 6>Control Flow </font>

**ControlFlow: Django is familiar with `index.html` coz in the views we have written path for `index.html` not for base.html,so the control reaches the `index.html` file,then it finds the first line as : `extends base.html`,so the control switches to base.html,there which traversing it finds the `{%block title}{%endblock title}`,after finding that the control shifts to again `index.html` and finds whether `{%block title}{%endblock title}` is present or not,if yes then what's the content in that block will be rendered!!,then the control see's the `{%block body}{%endblock body}`,in base.html,then the control shifts from base.html to again index.html to see whether `{%block title}{%endblock title}` is present or not!!,if yes,then whatever is the content in that block,i.e `{%block title}<content of the block>{%endblock title}`,that will be rendered!! and the control flows back to base.html and stops,and waits for the next request!!,so this is how Template inheritance in django works!!!**
